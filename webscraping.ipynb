```python
import csv
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.nfl.com/standings/league/2020/reg/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'lxml')
```

```python
#soup
```

```python
# extract the table
table = soup.find('table', {"class":"d3-o-table d3-o-table--row-striping d3-o-table--detailed d3-o-standings--detailed d3-o-table--sortable {sortlist: [[4,1]], sortinitialorder: 'desc'}"})
headers = []

for i in table.find_all('th'):
    title = i.text.strip()
    headers.append(title)
```

```python
df = pd.DataFrame(columns = headers)
df
```

```python
for row in table.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data
```

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.to_csv('Bella.csv', index=False)
```
